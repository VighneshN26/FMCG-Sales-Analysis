### Cleaning  Data

```python
df.isnull().sum()
```

```python
  #removing null values
df['Cylinders'].fillna(df['Cylinders'].mean(),inplace= True)
df['EngineSize'].fillna(df['EngineSize'].mean(),inplace= True)
df['Horsepower'].fillna(df['Horsepower'].mean(),inplace= True)
df['MPG_Highway'].fillna(df['MPG_Highway'].mean(),inplace= True)
df['Wheelbase'].fillna(df['Wheelbase'].mean(),inplace= True)  
df['MPG_City'].fillna(df['MPG_City'].mean(),inplace= True)
df['Weight'].fillna(df['Weight'].mean(),inplace= True)
df['Length'].fillna(df['Length'].mean(),inplace= True)
```

```python
df['MSRP'].fillna("Not Available",inplace =True)
df['Make'].fillna("Not Available",inplace =True)
df['Model'].fillna("Not Available",inplace =True)
df['Origin'].fillna("Not Available",inplace =True)
df['DriveTrain'].fillna("Not Available",inplace =True)
df['Invoice'].fillna("Not Available",inplace =True)
df['Type'].fillna("Not Available",inplace =True)
```
