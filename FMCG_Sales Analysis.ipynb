```python
import pandas as pd
import numpy as np
import re
```

```python
Data=pd.read_csv(r"C:\Users\vighnesh nirgude\Downloads\Raw Sales Data.csv")
```

```python
Data
```

```python
Data.info()
```

```python
#converting Data Type of "QTY" and "Sales" from object to float and removing comma's from value
Data["Sales"]=Data["Sales"].str.replace(",","").astype(float)
```

```python
Data["Qty"]=Data["Qty"].str.replace(",","").astype(float)
```

```python
#converting "month" to datetime formate"
Data["Month"]=pd.to_datetime(Data["Month"])
```

```python
#now changing "month (datetime format) " to "month name"(alphabetical format) and storing it into new column "month sales"
Data["Month sales"]=Data["Month"].dt.month_name(locale="english")
```

```python
#now removing numerical values from behind from product name for accurate calculation
Data["product"]=Data["Product Name"].apply(lambda x:re.split("(\d+)",x)[0])
```

```python
Data
```

### REQUIREMENT 1 = Monthwise number of quantities sold


```python
D1=pd.DataFrame(Data.groupby(Data["Month"])["Qty"].sum()).reset_index()
```

```python
D1
```

### REQUIREMENT 2 = Category wise Qty sold


```python
D2=pd.DataFrame(Data.groupby(Data["Category"])["Qty"].sum()).reset_index()
```

```python
D2
```

### REQUIREMENT 3 = Sub-channel wise total sales


```python
D3=pd.DataFrame(Data.groupby(Data["Sub-Channels"])["Qty"].sum()).reset_index()
```

```python
D3
```

### REQUIREMENT 4 = Highest sale in which month



```python
D4=pd.DataFrame(Data.groupby(Data["Month sales"])["Sales"].sum()).reset_index()
```

```python
D4.head(1)
```

# REQUIREMENT 5 = Highest selling product (based on sales) (packaging quantity doesn't matter. Ex. consider "GET SLIM
# JUICE 1 L" and "GET SLIM JUICE 500 ML" as one prduct

```python
D5=pd.DataFrame(Data.groupby(Data["product"]).sum().sort_values("Sales",ascending=False)).head()
```

```python
D5.head()
```

# REQUIREMENT 6 = List of different products company is offering (packaging quantity doesn't matter. Ex. consider "GET
# SLIM JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)


```python
D6=pd.DataFrame(Data.groupby(["Sub-Channels","product"]).sum()["Sales"])
```

```python
D6.sort_values(["Sub-Channels","Sales"],ascending=False)
```

```python
# Now saving the file in Excel file
writer = pd.ExcelWriter("Row_data_sales_analytics.xlsx", engine="xlsxwriter")
Data.to_excel(writer, sheet_name="Data",index=True)   
D1.to_excel(writer, sheet_name="Month Quantity Sold",index=True)
D2.to_excel(writer, sheet_name="Category QTY Sold",index=True)
D3.to_excel(writer, sheet_name="sub-channel sales",index=True)
D4.to_excel(writer, sheet_name="Month Highest sales",index=True)
D5.to_excel(writer, sheet_name="Highest selling product",index=True)
D6.to_excel(writer, sheet_name="Sub-Channel and product",index=True) 
writer.save()
Data
```
