```python
#importing liberaries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.figsize']=(12,8)
```

```python
df=pd.read_csv(r'C:\Users\vighnesh nirgude\Downloads\movies.csv')
```

```python
#looking Data
df.head()
```

```python
#cheacking any missing Data
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))

```

```python
#Data types of columns
df.dtypes
```

```python
#changing Datatype
df['budget']=df['budget'].astype('int64')
```

```python
#error occuring due to nan values in Dataset
#for correction using fillna meathod
df=df.fillna(0)
```

```python
#now changing Datatype
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df
```

```python
# creating correct year column
df['yearcorrect']=df['released'].str.extract(pat='([0-9]{4})')
df
```

```python
#highest grossing movies
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
#Drop any duplicates
df.drop_duplicates()
```

```python
#Budget high correlation
#company high correlation
```

```python
#scatter plot with budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Gross Earning ')
plt.ylabel('Budget of film')
plt.show()
```

```python
# plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "blue"}, line_kws={"color":"red"})
```

```python
# looking correlation 
df.corr(method='pearson') #pearson, kendall, spearman
```
